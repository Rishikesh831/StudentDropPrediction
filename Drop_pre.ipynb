```python
df = pd.read_csv('student_data.csv')
```

```python
df.head()
```

```python
df = df.drop('student_id',axis=1)
df.isnull().sum()
```

```python
df.info()
```

```python
from scipy.stats import ttest_ind

# Define your features (drop student_id and target)
features = df.drop(columns=[ "dropped_out"]).columns

# Split data into dropout = 1 and dropout = 0
dropout_group = df[df["dropped_out"] == 1]
stayed_group = df[df["dropped_out"] == 0]

results = []

for col in features:
    # Extract values for each group
    values_dropout = dropout_group[col]
    values_stayed = stayed_group[col]

    # Perform independent t-test
    t_stat, p_val = ttest_ind(values_dropout, values_stayed, equal_var=False)

    results.append({
        "Feature": col,
        "p_value": round(p_val, 4)
    })

# Convert to DataFrame for pretty output
pval_df = pd.DataFrame(results).sort_values(by="p_value")

print(pval_df)
```

```python
df["time_per_login"] = df["time_on_platform_hours"] / df["total_logins"]
df["time_per_login"].replace([float('inf'), -float('inf')], 0, inplace=True)
df["time_per_login"].fillna(0, inplace=True)


df["low_quiz_flag"] = (df["avg_quiz_score"] < 40).astype(int)


df["inactive_flag"] = (df["days_since_last_login"] > 15).astype(int)


df["assignments_per_login"] = df["%_assignments_completed"] / df["total_logins"]
df["assignments_per_login"].replace([float('inf'), -float('inf')], 0, inplace=True)
df["assignments_per_login"].fillna(0, inplace=True)
```

```python
df["engagement_index"] = (
    (df["total_logins"] * 1) +
    (df["%_assignments_completed"] * 1) +
    (df["time_on_platform_hours"] * 1) +
    (df["forum_posts"] * 3)
) / 6
```

```python
from scipy.stats import ttest_ind

# Define your features (drop student_id and target)
features = df.drop(columns=[ "dropped_out"]).columns

# Split data into dropout = 1 and dropout = 0
dropout_group = df[df["dropped_out"] == 1]
stayed_group = df[df["dropped_out"] == 0]

results = []

for col in features:
    # Extract values for each group
    values_dropout = dropout_group[col]
    values_stayed = stayed_group[col]

    # Perform independent t-test
    t_stat, p_val = ttest_ind(values_dropout, values_stayed, equal_var=False)

    results.append({
        "Feature": col,
        "p_value": round(p_val, 4)
    })

# Convert to DataFrame for pretty output
pval_df = pd.DataFrame(results).sort_values(by="p_value")

print(pval_df)
```
